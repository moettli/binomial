```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import pandas as pd

# Import data that should be fitted
df = pd.read_csv("Archer.csv", sep=";")

# x- und y-Werte auslesen
x = df["x"].values
y = df["y"].values

z = np.polyfit(x, y, 11)


# Generate y values using the fitted parameters
y_fit = np.poly1d(z)

# **Smooth x-grid**
x_smooth = np.linspace(min(x), max(x), 1000)
y_smooth = y_fit(x_smooth)

plt.scatter(x, y, label="Sample Data", color="green")
plt.plot(x_smooth, y_smooth, label="Fitted Curve", linewidth=2, color="red")

plt.xlabel("displacement / m")
plt.ylabel("force / N")
plt.title("Archer")
plt.legend()
plt.show()
```

Area: For calculating the area under the graph we integrate the polynomial from 0 to 0.5

```python
plt.plot(x_smooth, y_smooth, label="Fitted Curve", linewidth=2, color="red")

plt.xlabel("displacement / m")
plt.ylabel("force / N")
plt.title("Archer")
plt.legend()
#Fill under the curve
plt.fill_between(
        x_smooth, 
        y_fit(x_smooth), 
        color= "pink",
        alpha= 0.2)
plt.show()
```

```python
result, error = quad(y_fit,0,0.5)
```
