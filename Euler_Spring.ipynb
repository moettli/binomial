Using Python to Calculate an Oscillation
=====




# Load the libraries

```python
import matplotlib.pyplot as plt
```

## Define the parameters

```python
dt = 0.001
k = 10
m = 1
t0 = 0
v0 = 0
s0 = 0.1
a0 = -k*s0/m
tnew = t0
snew = s0
vnew = v0
anew = a0
```

## Initialize the lists

```python
t = [t0]
s = [s0]
v = [v0]
a = [a0]
```

## Calculate the next steps and append to lists


```python
for i in range(10000):
    tnew = tnew + dt
    snew = snew + vnew * dt + 1/2 * anew * dt * dt
    F = -k*snew
    anew = F/m
    vnew = vnew + anew * dt

    t.append(tnew)
    s.append(snew)
    v.append(vnew)
    a.append(anew)
```

## Plot the graphs

```python
plt.plot(t,s)
plt.plot(t,v)
plt.plot(t,a)
plt.grid()
plt.show()
```
